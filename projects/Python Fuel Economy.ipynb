```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style = "darkgrid")
mpg = sns.load_dataset("mpg")
```

```python
mpg
```

```python
# Create a visual showing the relationship between horsepower and fuel economy
sns.relplot(
    data = mpg,
    x = "horsepower",
    y = "mpg"
)
```

```python
#summarizing fuel economy by country of origin
sns.boxplot(
    data = mpg,
    x = "origin",
    y = "mpg"
)
```

```python
summary_table = mpg.groupby("origin")["mpg"].agg(
    mean_mpg = "mean",
    max_mpg = "max",
    min_mpg = "min"
).reset_index()


print(summary_table)
```

```python
#Showing how fuel economy has changed over time
sns.relplot(
    data = mpg,
    x = "model_year",
    y = "mpg",
    kind = "line"
)
```

```python
#Showing relationship between engine displacement and fuel economy
sns.relplot(
    data = mpg,
    x = "displacement",
    y = "mpg",
    hue = "cylinders",
    style = "origin",
    height = 6
)
```
